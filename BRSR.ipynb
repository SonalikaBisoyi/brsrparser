```python
import fitz  # PyMuPDF
import re

def fnPDF_FindText(search_string, start_page):
    for page_num in range(start_page, num_pages):
        page = reader.load_page(page_num)
        text = page.get_text()
        if re.search(search_string, text, re.IGNORECASE):  # Added re.IGNORECASE for case insensitive search
            print(f"String - '{search_string}', found on Page: {page_num}")
            return page_num
    return None

# Usage example (ensure you have the reader and num_pages correctly defined)
pdf_path = "/content/drive/My Drive/Colab Notebooks/riil-business-responsibility-and-sustainability-report-2022-23.pdf"
reader = fitz.open(pdf_path)
num_pages = reader.page_count

sections = [
    'SECTION A: GENERAL DISCLOSURES',
    'SECTION B: MANAGEMENT AND PROCESS DISCLOSURES',
    'Section C: Principle Wise Performance Disclosure'
]

section_pages = [0]

for i in range(len(sections)):
    start_search_string = sections[i]
    print('Start Search for section ' + start_search_string)

    # Find the start page for each section
    start_page = fnPDF_FindText(start_search_string, section_pages[-1])  # Use the last added page as start

    if start_page is not None:
        section_pages.append(start_page)
    else:
        print(f"Could not find section: {start_search_string}")
        # Optionally, handle this case (e.g., append a placeholder, break, or continue)

# Ensure the last page is added to section_pages if it is not already included
if num_pages not in section_pages:
    section_pages.append(num_pages)

# Remove duplicates and sort in ascending order
section_pages = sorted(set(section_pages))
print(section_pages)
```
